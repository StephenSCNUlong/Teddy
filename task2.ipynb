```python
#毛利润饼图
PROFIT_SUM = 0
ACC = []
for i in ['A', 'B', 'C', 'D', 'E']:

    data_i = data[data['地点'] == i]
    result_i = pd.merge(data_i, goods_class, how='left', on=['商品'])
    DRK = result_i[result_i['大类'] == '饮料']
    ANDRK = result_i[result_i['大类'] == '非饮料']
    DRK_prf = DRK['实际金额'].sum()
    ANDRK_prf = ANDRK['实际金额'].sum()
    
    PROFIT = DRK_prf*0.25 + ANDRK_prf*0.2
    PROFIT = round(PROFIT, 3)
    print(PROFIT)
    ACC.append(PROFIT)
    print(i, '售货机毛利润：', PROFIT)
    PROFIT_SUM = PROFIT_SUM + PROFIT

print(PROFIT_SUM)


PIE = (ACC/PROFIT_SUM).astype('float32')
print(PIE)
def draw_pie(labels, quants):
    plt.figure(1)
    
    expl = [0.1, 0, 0, 0, 0]
    colors = ['blue', 'red', 'coral', 'green', 'yellow']
    
    plt.pie(quants, explode=expl, colors=colors, labels=labels, autopct='%1.1f%%', pctdistance=0.8, shadow=True)
    plt.title('毛利润饼图')
    plt.savefig('毛利润饼图.png', dpi=300)
    plt.show()

    
MONTH = ['A', 'B', 'C', 'D', 'E']
draw_pie(MONTH, PIE)
```

```python
#气泡图,这里的月交易额均值计算方法为为：每月的交易额/该月天数
data_mer = pd.merge(data, goods_class, how='left', on=['商品'])

data_mer['支付时间'] = pd.to_datetime(data_mer['支付时间'])

#生成每月的日交易额均值
Ave_mon_pft = []
for i in L2class:
    Ave_mon_pft_i = []
    for j in range(1,13):
        data_me = data_mer[data_mer['支付时间'].dt.month == j]
        data_m = data_me[data_mer['二级类'] == i]
        nu, monthRange = calendar.monthrange(2017,j)
        re = data_m['实际金额'].sum() / monthRange
        Ave_mon_pft_i.append(re)
    
    Ave_mon_pft.append(Ave_mon_pft_i)

#列表转数组
Ave_mon_pft = np.array(Ave_mon_pft)
#保留3位小数
Ave_mon_pft = np.round(Ave_mon_pft, 3)
#print(Ave_mon_pft)
sns.set(style = "white")#设置样式
matplotlib.rcParams['font.sans-serif'] = ['SimHei'] 
matplotlib.rcParams['font.family'] = 'sans-serif'

matplotlib.rcParams['axes.unicode_minus'] = False
plt.figure(1)
plt.figure(1, figsize=(200,100))
cm = plt.cm.get_cmap('RdYlBu')
#这里我用0-19代替商品二级类作为纵轴，然后在替换数字为字符串
for i in range(0,len(L2class)):
        plt.scatter(range(1,13), np.array(i).repeat(12),s=Ave_mon_pft[i], c=Ave_mon_pft[i]  ,cmap=cm, linewidth = 0.5, marker='o' ,alpha = 0.5)
        
plt.yticks(range(0,20), L2class)
plt.xlabel('month')
plt.ylabel('二级类')
plt.colorbar()
plt.savefig('气泡图.png', dpi=300)
plt.show()
```

```python
#绘制售货机C6、7、8三个月订单量的热力图， 横轴以天为单位， 纵轴以小时为单位。

data_c = data[data['地点'] == 'C']
for i in [6, 7, 8]:
    data_c_JAS = data_c[ data_c['支付时间'].dt.month == i]
    data_c_need = data_c_JAS
    data_c_need['Month'] = data_c_need['支付时间'].dt.strftime('%b%d')
    data_c_need['Hour'] = data_c_need['支付时间'].dt.strftime('%H')
    data_c_ned = data_c_need[['支付时间', 'Month', 'Hour']]

    fina = data_c_ned.set_index(['Month', 'Hour'], drop=False)
    fina['cou'] = 1
    final = fina['cou']
    finalll = final.sum(level=[0,1])
    finaly = finalll.unstack(0).fillna(0)

    # plot heatmap
    ax = sns.heatmap(finaly, cmap='rainbow')

    # turn the axis label
    for item in ax.get_yticklabels():
        item.set_rotation(0)

    for item in ax.get_xticklabels():
        item.set_rotation(45)
        
    if i == 6:
        plt.savefig('6月热力图.png', dpi=300)
    if i == 7:
        plt.savefig('7月热力图.png', dpi=300)
    if i == 8:
        plt.savefig('8月热力图.png', dpi=300)
        
    plt.show()

```
